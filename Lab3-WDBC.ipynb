#### Import WDBC label data

```python
import pandas as pd
df = pd.read_csv('wdbc.data',header=None)
df
```

#### Printing the unique values in the column 1

```python
df.iloc[:,1].unique()
```

#### Giving labels to the Dataframe and Replacing M -> Malignant and B-> Benign in the Diagnosis column. Indicating which data corresponds to Malignant and Benign Cancer respectively

```python
column_labels = ['patient_id', 'diagnosis', 'mean radius', 'mean texture', 'mean perimeter', 'mean area',
'mean smoothness', 'mean compactness', 'mean concavity',
'mean concave points', 'mean symmetry', 'mean fractal dimension',
'radius error', 'texture error', 'perimeter error', 'area error',
'smoothness error', 'compactness error', 'concavity error',
'concave points error', 'symmetry error',
'fractal dimension error', 'worst radius', 'worst texture',
'worst perimeter', 'worst area', 'worst smoothness',
'worst compactness', 'worst concavity', 'worst concave points',
'worst symmetry', 'worst fractal dimension']
df.columns = column_labels

df['diagnosis'].replace('B','Benign',inplace=True)
df['diagnosis'].replace('M','Malignant',inplace=True)
```

```python
df
```

#### Spliting the Dataframe into two Feature Dataframe and Label Dataframe

```python
feature_df = df.iloc[:,2:]
label_df = df.iloc[:,1]
```

```python
label_df
```

```python
feature_df
```

#### Using StandardScaler for standard deviation to be 1

```python
from sklearn.preprocessing import StandardScaler

x = feature_df.values
x = StandardScaler().fit_transform(x)
```

#### Printing the Mean and Standard Deviation. Here the standard deviation is 0.1

```python
import numpy as np
np.mean(x),np.std(x)
```

#### Getting the Principal Component 1 and Principal Component 2 of the WDBC Training Dataset

```python
from sklearn.decomposition import PCA

pca_breast = PCA(n_components=3)
principal_Components_breast = pca_breast.fit_transform(x)
principal_breast_df = pd.DataFrame(data=principal_Components_breast,columns=['principal component 1','principal component 2','principal component 3'])
```

```python
pca_breast.explained_variance_ratio_
```

```python
principal_breast_df
```

#### Plotting the graph for Principal Component Analysis. The variance ratio of the Principal Component 1 is 0.44 and Principal Component 2 is 0.18. Hence the graph shows a clear separation of the data

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure()
plt.figure(figsize=(10,10))
plt.xlabel('PC 1')
plt.ylabel('PC 2')
targets = ['Benign','Malignant']
colors = ['b','r']

for targets,color in zip(targets,colors):
    indicesToKeep = label_df == targets
    plt.scatter(principal_breast_df.loc[indicesToKeep,'principal component 1'],principal_breast_df.loc[indicesToKeep,'principal component 2'], c=color)
    
plt.legend(targets,prop={'size':15})    
```

```python
pca = PCA().fit(x)
evr = pca.explained_variance_ratio_
print('\n'.join('{:2d}:{:0.4f}'.format(*k) for k in enumerate(evr)))
```

#### Plotting the Principal Component

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of components')
plt.ylabel('Total Explained Variance')
```
